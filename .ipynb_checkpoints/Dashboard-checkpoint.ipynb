```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
df = pd.read_excel('Elec_dashboard.xlsx', sheet_name='Sheet1')

# Convert 'Month & Year' to datetime
df['Month & Year'] = pd.to_datetime(df['Month & Year'])

# Select energy columns (excluding NaN columns for Nov-Dec 2024)
energy_columns = ['Total Stream DN Electricity - kWh', 'MTHW - kWh', 'Steam - kWh', 
                 'LPG - kWh', 'Woodchip/ Pellet - kWh', 'Solar - kWh']
df = df[df['Month & Year'] < '2024-11-01']

# Plot 1: Time Series of Energy Usage
plt.figure(figsize=(15, 8))
for column in energy_columns:
    plt.plot(df['Month & Year'], df[column], marker='o', label=column.split(' -')[0])

plt.title('Monthly Energy Consumption by Type (Jan 2022 - Oct 2024)')
plt.xlabel('Month-Year')
plt.ylabel('Consumption (kWh)')
plt.xticks(rotation=45)
plt.legend(bbox_to_anchor=(1.05, 1))
plt.grid(True)
plt.tight_layout()
plt.show()

# Plot 2: Energy Distribution
total_energy = df[energy_columns].sum()
plt.figure(figsize=(10, 8))
plt.pie(total_energy, labels=[col.split(' -')[0] for col in energy_columns], 
        autopct='%1.1f%%')
plt.title('Distribution of Energy Sources (2022-2024)')
plt.show()

# Plot 3: Monthly Average by Energy Type
monthly_avg = df.groupby(df['Month & Year'].dt.month)[energy_columns].mean()
plt.figure(figsize=(12, 6))
for column in energy_columns:
    plt.plot(monthly_avg.index, monthly_avg[column], 
             marker='o', label=column.split(' -')[0])

plt.title('Average Monthly Energy Consumption Pattern')
plt.xlabel('Month')
plt.ylabel('Average Consumption (kWh)')
plt.legend(bbox_to_anchor=(1.05, 1))
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate and display statistics
print("\nKey Statistics (in kWh):")
stats = df[energy_columns].describe().round(2)
display(stats)

# Calculate year-over-year changes
print("\nYear-over-Year Changes:")
for year in [2023, 2024]:
    prev_year = df[df['Month & Year'].dt.year == year-1][energy_columns].mean()
    curr_year = df[df['Month & Year'].dt.year == year][energy_columns].mean()
    change = ((curr_year - prev_year) / prev_year * 100).round(2)
    print(f"\n{year} vs {year-1} (% change):")
    for col, val in change.items():
        print(f"{col.split(' -')[0]}: {val}%")
```
